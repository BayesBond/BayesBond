```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.linalg import norm

results = {}

selectivity = np.linspace(0.01, 1, 100)

quality = selectivity
quantity = 100 * (1 - selectivity)
utility = quality * quantity


fig, ax = plt.subplots(3, 1, sharex=True)
ax[0].set_title("Qualitative explanation")
ax[0].plot(selectivity, quality)
ax[0].set_ylabel("Quality")
ax[1].plot(selectivity, quantity)
ax[1].set_ylabel("Quantity")
ax[2].plot(selectivity, utility)
ax[2].set_ylabel("App Benefit")
ax[2].set_xlabel("Selectivity")
plt.savefig("rational_qualitative.png")
plt.show()
```
